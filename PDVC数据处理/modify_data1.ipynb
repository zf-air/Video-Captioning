```python
# start to end 二维数组
start_to_end_numpy = np.zeros((len(mat_numpy),2))
```

```python
start_to_end_numpy
```

```python
for i in range(len(mat_numpy)):
    start_to_end_numpy[i][0] = mat_numpy[i][0]
    k = 6
    while(k>=0 and mat_numpy[i][k]==0):
        k = k - 1   
    if(k==0):
        start_to_end_numpy[i][1] = mat_numpy[i][0]
    else:
        start_to_end_numpy[i][1] = mat_numpy[i][k-1]
```

```python
start_to_end_numpy
```

```python
start_to_end_pandas = pd.DataFrame(start_to_end_numpy)
```

```python
for i in range(len(start_to_end_numpy)):
    start_to_end_numpy[i][0] = math.floor(start_to_end_numpy[i][0])
    start_to_end_numpy[i][1] = math.floor(start_to_end_numpy[i][1])
```

```python
num_frame_numpy = np.zeros((len(start_to_end_numpy),2))
```

```python
for i in range(len(start_to_end_numpy)):
    num_frame_numpy[i][0] = start_to_end_numpy[i][0] * 15
    num_frame_numpy[i][1] = start_to_end_numpy[i][1] * 15   
```
